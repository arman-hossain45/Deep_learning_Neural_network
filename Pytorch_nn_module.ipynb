```python
import torch.nn as nn
```

```python
class MySimpleNN(nn.Module):
  def __init__(self,num_features):
    super().__init__()
    self.linear=nn.Linear(num_features,1)
    self.sigmoid=nn.Sigmoid()

  def forward(self,features):
    out=self.linear(features)
    out=self.sigmoid(out)
    return out





```

# Here in only a One neuron

```python
import torch
import torch.nn as nn  # base class sob module ase aikhan neural network ar

class Model(nn.Module):  #  aikhan a amra nn class ke inherit kotesi karon ta na hole amra nn class ar sob
#functionality use korte parbo na. nn hol0 akta build in class
    def __init__(self, num_features):
        super().__init__()# we use super keyword child class parent class ar constructor ke call korte pare
        self.linear = nn.Linear(num_features, 1)  # Capital L, single neuron we create
        # Here input is num 0f feature and 1 output
        self.sigmoid = nn.Sigmoid()              # Capital S

    # forward pass
    def forward(self, features):
        out = self.linear(features)  # z = wx + b
        out = self.sigmoid(out)
        return out


# create a dummy dataset
features = torch.rand(10, 5)

# create a model
# object create of this model
model = Model(features.shape[1])

# call model for forward pass
output = model(features)
print(output)
```

# Here now we code With hidden layer

5 input and 3 neuron of 1 hidden layer  and 1 output layer

```python
class Model(nn.Module):  #  aikhan a amra nn class ke inherit kotesi karon ta na hole amra nn class ar sob
#functionality use korte parbo na. nn hol0 akta build in class
    def __init__(self, num_features):
        super().__init__()# we use super keyword child class parent class ar constructor ke call korte pare

        self.linear1 = nn.Linear(num_features, 3)  #here number of feature is 5 and hidden laayer 1 is 3 neuron
        #and one output layer which is connected in one hidden layer which has 3 neuron

        self.relu=nn.ReLU()# we use relu activation function in hidden layer

        self.linear2 = nn.Linear(3, 1)#use sigmoid in output layer

        self.sigmoid = nn.Sigmoid()#use sigmoid in output layer

    # forward pass
    def forward(self, features):
        out = self.linear1(features)  # z = wx + b
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)

        return out
```

```python
# create model

features = torch.rand(10, 5)
model = Model(features.shape[1])

model(features)
```

# Now in this code we do all of this things use a squential container

```python
class Model(nn.Module):  #  aikhan a amra nn class ke inherit kotesi karon ta na hole amra nn class ar sob
#functionality use korte parbo na. nn hol0 akta build in class
    def __init__(self, num_features):

        super().__init__()# we use super keyword child class parent class ar constructor ke call korte pare
# now we can do all of this using a sequential container
        self.network = nn.Sequential(
            nn.Linear(num_features,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )




    # forward pass
    def forward(self, features):
        out =self.network(features)


        return out
```

```python
# create model

features = torch.rand(10, 5)
model = Model(features.shape[1])

model(features)
```
