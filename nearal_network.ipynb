```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.datasets import mnist
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
```

Data set loading

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

#we want to see multiple  image at a time so that we use a loop to see the all the image

```python
plt.figure(figsize=(10,10))# we see this include with label
for i in range(25):
  plt.subplot(5,5,i+1)
  plt.imshow(x_train[i],cmap="gray")
  plt.title(f"Label : {y_train[i]}")

plt.tight_layout()
```

# Normalze the data reshape the data our data is 28*28 shape

```python
x_train=x_train.reshape(-1,28*28)/255.0  # Normalization ensure faster and more stable  training by scaling input values
x_test=x_test.reshape(-1,28*28)/255.0   # pixel value maximum range is 255
```

# define the Neural network Architecture

```python
model=Sequential(
    [
        Dense(128 , activation='relu', input_shape=(28*28,)), # input layer
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),# hidden layer
        Dense(10,activation='softmax')# output layer
    ]
)
```

# complile the model

```python
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=['accuracy']
)
```

# Train the model

মানে:

একবারে 32টা image নেবে

prediction করবে

loss হিসাব করবে

weight update করবে

1407 image / 32 ≈ 1406 batch


```python
model.fit(
    x_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_split=0.25
)
```

# evaluate the models performance on test data

```python
test_loss,test_accuracy=model.evaluate(x_test,y_test)
print(f"\n Test Accuracy : { test_accuracy}")
print(f"\n Test Loss : {test_loss}")
```

# predict on the  test set

```python
y_pred=model.predict(x_test)
y_pred
```

# display the confusion matrix

```python
#convert probability to class label
y_pred_classes=np.argmax(y_pred,axis=1)# convert probabilities into class label
y_pred_classes
```

```python
cm=confusion_matrix(y_test,y_pred_classes)
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=np.arange(10))
disp.plot(cmap=plt.cm.Blues)
```

# save the model

```python

model.save('nn_model.h5')
```

# Prediction

```python
random_index = np.random.randint(0, len(x_test))
random_image = x_test[random_index]
random_label = y_test[random_index]

random_image_reshaped = random_image.reshape(1, 28 * 28)
predictions = loaded_model.predict(random_image_reshaped)
predicted_label = np.argmax(predictions)

print(f"True Label: {random_label}")
print(f"Predicted Label: {predicted_label}")

plt.imshow(random_image.reshape(28, 28), cmap='gray')
plt.title(f"True Label: {random_label}, Predicted: {predicted_label}")
plt.axis('off')
plt.show()
```
