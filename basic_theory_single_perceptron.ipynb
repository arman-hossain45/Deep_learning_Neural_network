## Write Answer 05:

A single-layer perceptron fails to solve the XOR problem because XOR is not linearly separable.

Linear models are not always sufficient

Complex patterns

Non-linear relationships

A single-layer perceptron can’t capture these.


A multilayer neural network with hidden layers:

Combines multiple linear boundaries

Applies non-linear activation functions

Effectively creates non-linear decision boundaries

# Question 06: [ Marks 20 ]

What is meant by the “perceptron Learning Rule” or weight adjustment method?
Why can adjusting one weight affect the classification of other points?

## Write Answer 06:

The perceptron learning rule is an error-driven method used to adjust the weights and bias of a perceptron so that it correctly classifies training data.

When the perceptron makes a mistake, it changes its weights to reduce that mistake next time.

w(new)=w(old)+eta*(y-y_pred)*x

b(new)=b(old)+eta*(y-y_pred)

here weights are updated only if the prediction is wrong.


Changing one weight:

Rotates the decision boundary

Changing the bias:

Shifts the boundary

Since the same boundary is used for all data points, any adjustment affects every point, not just the one that caused
the update.


so,the perceptron learning rule updates weights to correct errors, and since weights control a single shared decision boundary, changing even one weight can alter the classification of many points.
